```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
df = pd.read_csv("city_day.csv")
df.head()
```

```python
df.shape
```

```python
df.describe().transpose()
```

```python
df[['year','month','day']]=df['Date'].str.split('-',expand=True)
```

```python
df.head()
```

```python
df.drop("Date",axis=1,inplace=True)
```

```python
df.head()
```

```python
data=df.copy()
```

```python
data.head()
```

```python
data.shape
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data = data.drop(["PM10","NH3","Xylene"],axis=1)
data.head()
```

```python
data.dropna(inplace=True)
data.head()
```

```python
data.shape
```

```python
from sklearn.preprocessing import OrdinalEncoder
a = OrdinalEncoder()
data[["city"]] = a.fit_transform(data[["City"]])
data[["Remark"]] = a.fit_transform(data[["AQI_Bucket"]])
data[["City","city","AQI_Bucket","Remark"]]
```

```python
data["City"].unique()
```

```python
data["day"] = data["day"].astype("int64")
data["month"] = data["month"].astype("int64")
data["year"] = data["year"].astype("int64")
```

```python
data.info()
```

```python
data
```

```python
from sklearn.linear_model import LinearRegression,LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,r2_score
```

```python
data1=data.drop(["City","AQI_Bucket","Remark"],axis=1)
train=data1.drop("AQI",axis=1)
test=data1["AQI"]
lr=LinearRegression().fit(train,test)
r2_score(test,lr.predict(train))
```

```python
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators = 50)
clf.fit(train,test)
```

```python
r2_score(test,lr.predict(train))
```

```python
x = data[["AQI"]]
y = data[["Remark"]]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=42)


rem = RandomForestClassifier(n_estimators = 5).fit(x,y)
```

```python
r2_score(y,rem.predict(x))
```

```python
from tkinter import *
from PIL import Image, ImageTk

root = Tk()
root.title("Air Quality Index")
def airinfo():
    try:
        click = option.index(city.get())
        monthdata=monthopt.index(month.get())+1
        print(entrypm2.get())
        res = clf.predict([[entrypm2.get(),entryno.get(),entryno2.get(),entrynox.get(),entryco.get(),entryso2.get(),entryo3.get(),entryben.get(),entrytol.get(),entryyear.get(),monthdata,entryday.get(),click]])
        res1=res.reshape(1,1)
        print(res)
        remark2 = rem.predict(res1)
        remark2=math.ceil(remark2)
        print(remark2)
        if(remark2 == 0):
            remark1="Good"
            impact="Minimal impact"
            label1.config(text=remark1,fg='green')
            label2.config(text=impact,fg='green')
    
        elif remark2 == 3:
            remark1="Satisfactory"
            impact="Minor breathing discomfort to sensitive people"
            label1.config(text=remark1)
            label2.config(text=impact,fg='red')
    
        elif remark2 == 1:
            remark1="Moderate"
            impact="Breathing discomfort to the people with lungs, asthma ans heart diseases"
            label1.config(text=remark1)
            label2.config(text=impact,fg='red')
    
        elif remark2 == 5:
            remark1="Very Poor"
            impact="Respiratory illness on prolonged exposure"
            label1.config(text=remark1)
            label2.config(text=impact,fg='red')
    
        elif remark2 == 2:
            remark1="Poor"
            impact="Breathing discomfort to most people"
            label1.config(text=remark1)
            label2.config(text=impact,fg='red')
        else:
            remark1="Severe"
            impact="Affect healthy people and seriously impacts those with existing diseases"
            label1.config(text=remark1)
            label2.config(text=impact,fg='red')
    
    except ValueError:
        remark1="Invalid Input"
        impact="Please Fill the boxes carefully"
        label1.config(text=remark1)
        label2.config(text=impact,fg='red')
        
def cleanbox():
    
    entrypm2.delete(0,'end')
    entryno.delete(0,'end')
    entryno2.delete(0,'end')
    entrynox.delete(0,'end')
    entryco.delete(0,'end')
    entryso2.delete(0,'end')
    entryo3.delete(0,'end')
    entryben.delete(0,'end')
    entrytol.delete(0,'end')
    entryyear.delete(0,'end')
    entryday.delete(0,'end')
    city.set("Ahmedabad")
    month.set("January")
    label1.config(text="")
    label2.config(text="")
    
#geometry
root.geometry("1000x667")
#Image
image = Image.open("background.jpg")
photo = ImageTk.PhotoImage(image)
bgd = Label(image=photo)
bgd.place(x=0,y=0)

#Heading
head=Label(root,text="Air Quality Index", font="Arial 15 bold",pady=10,highlightthickness=2,bg="#050A30",fg="White")

head.grid(row=0,column=3)



#creating Labels
Label(root,text="City",pady=10,bg="#050A30",fg="white",width=7).grid(row=1,column=0)
Label(root,text="PM2.5",pady=10,bg="#050A30",fg="white",width=7).grid(row=2,column=0)
Label(root,text="NO",pady=10,bg="#050A30",fg="white",width=7).grid(row=3,column=0)
Label(root,text="NO2",pady=10,bg="#050A30",fg="white",width=7).grid(row=4,column=0)
Label(root,text="NOx",pady=10,bg="#050A30",fg="white",width=7).grid(row=5,column=0)
Label(root,text="CO",pady=10,bg="#050A30",fg="white",width=7).grid(row=6,column=0)
Label(root,text="SO2",pady=10,bg="#050A30",fg="white",width=7).grid(row=7,column=0)
Label(root,text="O3",pady=10,bg="#050A30",fg="white",width=7).grid(row=8,column=0)
Label(root,text="Benzene",pady=10,bg="#050A30",fg="white",width=7).grid(row=9,column=0)
Label(root,text="Toluene",pady=10,bg="#050A30",fg="white",width=7).grid(row=10,column=0)
Label(root,text="Year",pady=10,bg="#050A30",fg="white",width=7).grid(row=11,column=0)
Label(root,text="Month",pady=10,bg="#050A30",fg="white",width=7).grid(row=12,column=0)
Label(root,text="Day",pady=10,bg="#050A30",fg="white",width=7).grid(row=13,column=0)
  
Label(root,text="Result : ",font="bold",bg="#050A30",fg="white",width=10).grid(row=21,column=0,sticky=W)

label1=Label(root,text="",font="bold")
label1.grid(row=20,column=3)

label2=Label(root,text="",font="bold")
label2.grid(row=21,column=3)

#create variable

city= StringVar()
pm2 = IntVar()
no = IntVar()
no2 = IntVar()
nox = IntVar()
co = IntVar()
so2 = IntVar()
o3 = IntVar()
ben = IntVar()
tol = IntVar()
year = IntVar()
month = StringVar()
day = IntVar()

#textbox:
entrypm2 = Entry(root,textvariable=pm2,text="",highlightthickness=2)
entryno = Entry(root,textvariable=no,text="",highlightthickness=2)
entryno2 = Entry(root,textvariable=no2,text="",highlightthickness=2)
entrynox = Entry(root,textvariable=nox,text="",highlightthickness=2)
entryco = Entry(root,textvariable=co,text="",highlightthickness=2)
entryso2 = Entry(root,textvariable=so2,text="",highlightthickness=2)
entryo3 = Entry(root,textvariable=o3,text="",highlightthickness=2)
entryben = Entry(root,textvariable=ben,text="",highlightthickness=2)
entrytol = Entry(root,textvariable=tol,text="",highlightthickness=2)
entryyear = Entry(root,textvariable=year,text="",highlightthickness=2)
entryday = Entry(root,textvariable=day,text="",highlightthickness=2)

#highlight boder

entrypm2.config(highlightbackground="#050A30")
entryno.config(highlightbackground="#050A30")
entryno2.config(highlightbackground="#050A30") 
entrynox.config(highlightbackground="#050A30") 
entryco.config(highlightbackground="#050A30")
entryso2.config(highlightbackground="#050A30") 
entryo3.config(highlightbackground="#050A30")
entryben.config(highlightbackground="#050A30") 
entrytol.config(highlightbackground="#050A30") 
entryyear.config(highlightbackground="#050A30")
entryday.config(highlightbackground="#050A30") 




entrypm2 .grid(row=2,column=2,sticky=W)
entryno.grid(row=3,column=2,sticky=W)
entryno2.grid(row=4,column=2,sticky=W)
entrynox .grid(row=5,column=2,sticky=W)
entryco.grid(row=6,column=2,sticky=W)
entryso2 .grid(row=7,column=2,sticky=W)
entryo3.grid(row=8,column=2,sticky=W)
entryben .grid(row=9,column=2,sticky=W)
entrytol.grid(row=10,column=2,sticky=W)
entryyear .grid(row=11,column=2,sticky=W)
entryday.grid(row=13,column=2,sticky=W)


#dropdown
option = ['Ahmedabad', 'Aizawl', 'Amaravati', 'Amritsar', 'Bengaluru',
      'Chandigarh', 'Chennai', 'Coimbatore', 'Delhi', 'Gurugram',
       'Hyderabad', 'Jaipur', 'Kolkata', 'Lucknow', 'Patna', 'Shillong',
       'Talcher', 'Visakhapatnam']

monthopt= ["January","Febuary","March","April","May","June","July","August","September","October","November","December"]

city.set("Ahmedabad")
month.set("January")

drop = OptionMenu(root, city, *option)
drop.grid(row=1,column=2)

monthdrop=OptionMenu(root,month,*monthopt)
monthdrop.grid(row=12,column=2)

    
#button
Button(text="Submit", command=airinfo,font="Arial 12 bold",width=10,fg="#050A30").grid(row=7,column=5,sticky=W)
Button(text="Clear", command=cleanbox,font="Arial 12 bold",width=10,fg="#050A30").grid(row=3,column=5,sticky=W)
Button(text="Exit", command=root.destroy,font="Arial 12 bold",width=10,fg="#050A30").grid(row=5,column=5,sticky=W)
root.mainloop() 
```
